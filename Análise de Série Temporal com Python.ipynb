```python
# definindo um vetor para armazenar uma série histórica
count = 25
#print("Minha variável:", count)

anos_observados = [0] * count
#print(anos_observados)
ano_inicial = 2000

# preenche o vetor que armazena a séria histórica
for x in range(count):
  anos_observados [x] = ano_inicial + x

#print(anos_observados)

# seleciona no vetor os anos desejados
anos_selecionados = [0] * 15
#print(anos_selecionados)

# preenche o vetor dos anos selecionados
i = 0
for x in anos_observados:
  if x >= 2010:
    anos_selecionados[i] = x
    i = i + 1

# preenche o valor de valores que eu desejo observar na série histórica
valores_seleciondos = [7.5, 4.0, 1.9, 3.0, 0.5, -3.5, -3.3, 1.3, 1.8, 1.1, -4.1, 4.6, 3.1, 3.2, 3.5]

# verifica os valores criados para a série histórica
print(anos_selecionados)
print(valores_seleciondos)
```

```python
# vizualização dos dados da série histórica
import pandas as pd
from plotnine import ggplot, aes, geom_line, ggtitle

# criar a série histórica utilizando uma estrutura de dados
df_dados = pd.DataFrame({"Ano": anos_selecionados, "PIB": valores_seleciondos})
print(df_dados)

#plotar um gráfico de tendência
ggplot(df_dados) + aes(x = "Ano", y= "PIB") + geom_line() + ggtitle("Série histórica do PIB brasileiro dos últimos 15 anos")

```
