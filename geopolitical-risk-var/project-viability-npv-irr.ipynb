```python
## Parâmetros do problema
# valor presente líquido (vpl)
import numpy_financial as npf

# dados iniciais do problema
valor_inicial = -350000 # investimento inicial do projeto
valor_entrada_saida = 90000 - 15000

# contrói o vetor de fluxo de caixa
vetor_fluxo_caixa = [valor_inicial,        # aporte inicial
                     valor_entrada_saida,  # desembolsos de fluco de caixa líquido (1-6)
                     valor_entrada_saida,  # sendo igual ao seu período n
                     valor_entrada_saida,
                     valor_entrada_saida,
                     valor_entrada_saida,
                     valor_entrada_saida]

print(f"Fluxo de caixa analisado: {vetor_fluxo_caixa}")

# calcula o valor presente líquido por meio da função
# npv() da biblioteca numpy_financial
vl_presente_liquido = npf.npv(rate=0.15, values=vetor_fluxo_caixa)

print(f"O valor Presente Líquido (VPL) é: {round(vl_presente_liquido, 2)}")

```

```python
# cálculo da TIR (Taxa Interna de Retorno)
import numpy_financial as npf
import numpy as np

# Parâmetros do problema
vl_aporte_inicial = -350000 - 15000
vetor_fluxo_caixa = np.zeros(7)
vetor_fluxo_caixa[0] = vl_aporte_inicial

# loop para preencher o vetor de fluxo de caixa
for i in range(6):
  vetor_fluxo_caixa[i+1] = valor_entrada_saida

vetor_fluxo_caixa

# calcular a TIR por meio da biblioteca numpy_financial
vl_calculado_tir = npf.irr(values=vetor_fluxo_caixa)
print("TIR = " + str(round(vl_calculado_tir*100, ndigits=2)) + "%")

```

```python
# plotar o gráfico da análise TIR
import matplotlib.pyplot as plt
import numpy as np

eixo_x = np.linspace(0, 0.35, 200)
VPL = np.array([npf.npv(i, vetor_fluxo_caixa) for i in eixo_x])

# construção do gráfico
print("Gabriela Alvim De Nigris (RA: 202203856219 - 30/03)")
plt.plot(eixo_x, VPL, '--k', linewidth=2)
plt.plot(vl_calculado_tir, 0, 'k*', markersize=15)
plt.xlabel('Juros - i')
plt.title("TIR - Análise da Taxa Interna de Retorno")
plt.grid()

```
