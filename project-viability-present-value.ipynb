```python
## Parâmetros do problema
# valor presente (VP)
# valor futuro (VF)
# taxa de juros de referência (r)
# número de períodos (n)

# Cálculo do valor presente
def calcula_valor_presente(vl_futuro, tx_juros, nr_periodos):
  return vl_futuro / (1+tx_juros)**nr_periodos
```

```python
# Teste da função VP (valor presente)
# Exemplo: quero ter 15.000 em um ano e a taxa de juros está 12,25% a.a.

vl_presente_calculado = calcula_valor_presente(vl_futuro = 15000,
                                                   tx_juros = 0.1225,
                                                   nr_periodos = 1)

print(f"O valor presente deve ser: {round(vl_presente_calculado,4)}")
```

```python
# Simulação de cálculo de viabilidade de projeto

# API do C=Bacen
# Baixar dataset
#
vetor_tx_juros = [12.90, 11.65, 10.15, 9.15, 8.15]

NUMERO_PERIODOS = 2
VL_FUTURO = 15000
print("Gabriela Alvim De Nigris (RA: 202203856219 - 23/03)")

for tx in vetor_tx_juros:
  vl_calculado = calcula_valor_presente(VL_FUTURO, tx, NUMERO_PERIODOS)
  print(f"O valor presente da taxa {tx} foi: {round(vl_presente_calculado, 4)}")
```
